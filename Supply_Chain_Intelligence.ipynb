```python
drop_cols = [
    'Customer Email','Customer Password','Product Description',
    'Product Image','Customer Fname','Customer Lname',
    'customer_street','customer_zipcode','order_zipcode',
    'latitude','longitude','product_status',
    'product_card_id','order_item_cardprod_id','order_item_id',
    'category_id','department_id','product_category_id'
]

df = df.drop(columns=drop_cols, errors='ignore')
```

```python
df.shape
```

```python
df.columns = df.columns.str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '')
```

```python
df.info()
```

```python
df['order_date'] = pd.to_datetime(df['order_date_dateorders'])
df['shipping_date'] = pd.to_datetime(df['shipping_date_dateorders'])
```

```python
df[['order_date', 'shipping_date']].dtypes
```

```python
df['delivery_gap'] = df['days_for_shipping_real'] - df['days_for_shipment_scheduled']
```

```python
df.info()
```

```python
def delivery_performance(x):
    if x > 2:
        return 'Severely Late'
    elif x > 0:
        return 'Late'
    elif x == 0:
        return 'On Time'
    else:
        return 'Early'

df['delivery_performance'] = df['delivery_gap'].apply(delivery_performance)
```

```python
threshold = df['sales'].quantile(0.75)

df['high_value_order'] = (df['sales'] > threshold).astype(int)
```

```python
df['low_profit_flag'] = (df['order_profit_per_order'] < 0).astype(int)
```

```python
demand = (
    df.groupby(['product_name', 'order_date'])['order_item_quantity']
    .sum()
    .reset_index()
    .sort_values(['product_name', 'order_date'])
)
```

```python
demand['inventory'] = 500  # assumption

for i in range(1, len(demand)):
    if demand.loc[i, 'product_name'] == demand.loc[i-1, 'product_name']:
        demand.loc[i, 'inventory'] = max(
            demand.loc[i-1, 'inventory'] - demand.loc[i, 'order_item_quantity'],
            0
        )
```

```python
demand['stockout_flag'] = (demand['inventory'] == 0).astype(int)
```
